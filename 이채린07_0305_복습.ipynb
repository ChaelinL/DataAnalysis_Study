# 데이터 분석 기초반 첫 수업
- 데이터프레임에 대한 이해가 완벽해야 한다.
- 데이터 프레임((df)
- 2차원 / 2차원 리스트

```python
a = [[1,2],[3,4],[5,6]]
a
```

```python
# 2차원 리스트의 요소를 접근하기 위해서 index를 기준으로 접근
a[0][0]  # 2차원
```

- x, y 좌표평면 2차원
- 데이터프레임은 행과 열로 행은 index, 열은 column
- 데이터 분석 전처리에 판다스는 필수, numpy 기본

```python
import pandas as pd

df = pd.DataFrame({'이름':['홍길동', '김명희', '김철수'],
'성적':['A', 'B', 'C'],
'점수':[10, 20, 30]})

df
```

- 2차원이라는 형태
- 컬럼, 열, 이름, 성적, 점수
- 0, 1, 2 index 개념은 열
- 데이터프레임은 행과 열로 이루어졌고, 데이터프레임을 추출하거나 조작할 때는 행과 열로 접근해야 한다!

```python
# 성적만 추출하기
df['성적']  # 시리즈 형태로 바뀜
# 시리즈 형태로 다양한 연산이 가능하고, numpy에서도 제공하는 것을 이용할 수 있다.
# 연산하기 가장 좋은 형태 판다스 제공한다.
```

```python
# 성적과 이름 추출하기
df[['이름', '성적']]
```

```python
# 인덱스를 기준으로 추출할 수 있다
df['점수'].sum()
```

```python
# mpg 데이터로 살펴보기!
from google.colab import drive
drive.mount('/content/drive')

mpg = pd.read_csv('/content/drive/My Drive/mpg.csv')
mpg
```

```python
# 이때, 컬럼을 기준으로 데이터를 추출했는데 model, hwy는 데이터 타입이 다르다.
mpg.info()
```

- object(문자열) 데이터는 어떤 식으로 접근해야 할까?
- 문자열 데이터를 숫자로 변경해야 한다. - 빈도수 or 라벨링(인코딩)

```python
# 아우디 차량은 몇 개가 있을까?
mpg['manufacturer'].value_counts()
```

```python
# 파생변수 만들기
mpg['tot'] = mpg['cty'] + mpg['hwy']
mpg
```

```python
import numpy as np

# tot를 기준으로 연비 우수/비우수 구분하기
mpg['test'] = np.where(mpg['tot']>40, 'good', 'bad')

mpg['test'].value_counts()
```

```python
# 그래프로 연비 우수/비우수 비교해보기
mpg['test'].value_counts().plot.bar()
```
