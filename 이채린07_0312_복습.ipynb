**query의 사용**
- query()행의 데이터를 추출하는 것 = 내가 원하는 데이터를 가져올 수 있다.
- query(조건식) => 조건식에 맞는 데이터가 나온다!

```python
exam.query('nclass==1')  # 1반인 학생들을 불러온다
```

```python
exam.query('nclass!=1')  # 1반이 아닌 학생들을 불러온다
```

```python
# 비교연산자
exam.query('math>=50')  # 수학점수가 50점 이상인 학생들을 불러온다
```

```python
# 논리연산자
exam.query('nclass==1 & math>=50')  # 1반이면서 수학점수가 50점 이상인 학생들을 불러온다
```

```python
exam.query('math>=50 | science>=50')  # 수학점수가 50점 이상이거나 과학점수가 50점 이상인 학생들을 불러온다
```

**groupby, agg 이용하기**
- 엑셀에서 말하는 피벗 개념
- 피벗이라는 건 전체 데이터를 가지고 내가 원하는 컬럼만 요약통계를 볼 수 있는 것
- 즉, 집단별로 요약통계를 구한다.
- groupby: 집단을 만드는 함수
- agg: 다양한 통계치를 사용할 수 있는 함수

```python
exam.groupby('nclass').agg(mean_english=('english','mean'))  # 파생변수이름, 사용할 컬럼(변수), 원하는 요약통계치
```

```python
exam.groupby('nclass').agg(sum_english=('english', 'sum'))
```

```python
exam.groupby('nclass').agg(med_english=('english', 'median'))  # 이처럼 max, min, std도 가능!
```

```python
# 여러 개를 요약통계치로 만들어보자!
exam.groupby('nclass').agg(mean_math=('math','mean'),
                           sum_math=('math','sum'),
                           max_math=('math','max'),
                           n = ('nclass','count'))
```

**merge**
- 데이터 병합
- 현업에서 두 개의 데이터프레임을 추출해 하나로 합치는 경우 多
- 데이터를 합친다는 것은 공통된 기준이 있어야 한다.

```python
a = pd.DataFrame({'name':['김철수', '김영희', '홍길동'],
                  '영어점수':[10, 20, 30]})
b = pd.DataFrame({'name':['김철수', '김영희', '홍길동'],
                  '수학점수':[10, 20, 40]})

display(a)
display(b)
```

```python
# 공통의 기준인 name을 이용해 merge 사용
tot = pd.merge(a, b, how='inner')
tot
```

**concat**
- 덩어리 + 덩어리로 붙는 형태

```python
pd.concat([a,b])  # 축의 디폴트는 0
```

```python
pd.concat([a,b], axis=1)
```

- merge와 concat은 어떤 상황에서 사용하느냐에 따라 다르다.
- 두 개의 원리를 정확히 이해하고 있어야 대량의 데이터를 합칠 때 문제가 발생하지 않는다.
