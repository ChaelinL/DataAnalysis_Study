# 데이터분석 기초반 6주차 과제1
- 도수분포표의 개념 설명과 예시 코드를 통해 시각화하기

**도수분포표**: 주어진 자료를 몇 개의 계급으로 나누고 각 계급의 도수를 조사하여 나타낸 표


**계급**: 변량을 일정한 간격으로 나눈 구간
- 계급의 크기: 구간의 너비, 즉 계급의 양 끝 값의 차
- 계급의 개수: 변량을 나눈 구간의 수


**도수**: 각 계급에 속하는 자료의 수

```python
# hist()함수를 이용한 도수분포표 시각화

import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(3,2,100)  # 랜덤 데이터 생성 (평균, 편차, 크기)
plt.hist(x)  # 히스토그램 생성
```

```python
# bins= 설정을 통한 계급의 개수 조정
plt.hist(x, bins=5)
```

```python
plt.hist(x, bins=10)  # bins 개수에 따라 y축의 scale 또한 자동으로 바뀌는 것을 알 수 있다.
```

- 시각화 특성상 데이터의 분포에 대한 왜곡을 시킬 수 있기 때문에, 이를 유의하여 bins를 신중하게 선택해야 한다. 
- 보통 bins의 개수는 15개~20개 미만으로 설정한다.
