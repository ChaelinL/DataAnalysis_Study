# 데이터분석 기초반 5주차 과제2
- 범례값 만들어 비교해보기! 
- best
- center, center left, center right
- right
- upper left, right, center
- lower left, right, center

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 산점도를 그려보자
plt.scatter([1,2,3],[4,5,6])
plt.scatter([5,6,7],[10,5,6])
plt.show()
```

**1. 범례 생성**
- 먼저 레이블을 지정해줘야 범례 생성이 된다!

```python
plt.scatter([1,2,3],[4,5,6], label='mpg.cty')
plt.scatter([5,6,7],[10,5,6], label='mpg.hwy')
plt.legend()  # 디폴트 설정값은 legend(loc='best')
              # 그래프의 모양을 보고 최적의 위치에 자동으로 범례 위치를 정한다.
plt.show()
```

**2. 범례 위치 지정**
- legend(loc='')를 통해서 범례의 위치를 조절할 수 있다.
- 총 11개의 범례 위치 설정 존재

```python
plt.scatter([1,2,3],[4,5,6], label='mpg.cty')
plt.scatter([5,6,7],[10,5,6], label='mpg.hwy')
plt.legend(loc='best')  # 디폴트 설정값이다.
plt.show()
```

```python
plt.scatter([1,2,3],[4,5,6], label='mpg.cty')
plt.scatter([5,6,7],[10,5,6], label='mpg.hwy')
plt.legend(loc='center')  # 그래프 중앙으로 위치를 설정한다.
plt.show()
```

```python
plt.scatter([1,2,3],[4,5,6], label='mpg.cty')
plt.scatter([5,6,7],[10,5,6], label='mpg.hwy')
plt.legend(loc='center left')  # 그래프 중앙의 왼쪽에 위치를 설정한다.
plt.show()
```

```python
plt.scatter([1,2,3],[4,5,6], label='mpg.cty')
plt.scatter([5,6,7],[10,5,6], label='mpg.hwy')
plt.legend(loc='center right')  # 그래프 중앙의 오른쪽에 위치를 설정한다.
plt.show()
```

```python
plt.scatter([1,2,3],[4,5,6], label='mpg.cty')
plt.scatter([5,6,7],[10,5,6], label='mpg.hwy')
plt.legend(loc='right')  # center right와 동일하다.
plt.show()
```

```python
plt.scatter([1,2,3],[4,5,6], label='mpg.cty')
plt.scatter([5,6,7],[10,5,6], label='mpg.hwy')
plt.legend(loc='upper left')  # 그래프 왼쪽 상단에 위치를 설정한다.
plt.show()
```

```python
plt.scatter([1,2,3],[4,5,6], label='mpg.cty')
plt.scatter([5,6,7],[10,5,6], label='mpg.hwy')
plt.legend(loc='upper right')  # 그래프 오른쪽 상단에 위치를 설정한다.
plt.show()
```

```python
plt.scatter([1,2,3],[4,5,6], label='mpg.cty')
plt.scatter([5,6,7],[10,5,6], label='mpg.hwy')
plt.legend(loc='upper center')  # 그래프 중앙 상단에 위치를 설정한다.
plt.show()
```

```python
plt.scatter([1,2,3],[4,5,6], label='mpg.cty')
plt.scatter([5,6,7],[10,5,6], label='mpg.hwy')
plt.legend(loc='lower left')  # 그래프 왼쪽 하단에 위치를 설정한다.
plt.show()
```
