```python
import pandas as pd

# crime 데이터 불러오기
from google.colab import drive
drive.mount('/content/drive')

crime = pd.read_csv('/content/drive/My Drive/crime.csv')
crime
```

```python
crime.info()
```

```python
# 시계열 데이터로 변경
crime['REPORTED_DATE'] = pd.to_datetime(crime['REPORTED_DATE'])
crime.info()
```

```python
crime_dt = crime.set_index('REPORTED_DATE')  # 시계열 컬럼을 인덱스로 지정
crime_dt = crime_dt.sort_index()
```

**1. resample 함수**
- 시계열 데이터를 다른 주기로 리샘플링
- 연, 월, 주, 일 선택 가능

```python
crime_dt .resample('1W').sum()  # 1주일 단위로 데이터를 리샘플링하고, 각 주에 해당하는 값을 합산한다
```

**2. first 함수**
- 시계열 데이터의 첫 번째 항목을 선택하는데 사용
- first를 사용하여 원하는 시작 시간을 지정할 수 있다.
- 연, 월, 주, 일 선택 가능

```python
crime_dt.first('5D')  # crime_dt 데이터셋의 항목을 앞에서부터 5일 단위로 불러온다.
```

**3. last 함수**
- 시계열 데이터의 마지막 항목을 선택하는데 사용
- 연, 월, 주, 일 선택 가능

```python
crime_dt.last('1W')  # crime_dt 데이터셋의 항목을 뒤에서부터 일주일 단위로 불러온다.
```

**4. shift 함수**
- 시계열 데이터를 주어진 오프셋만큼 이동

```python
crime_dt
```

```python
crime_dt.shift(2, freq='H')  # crime_dt의 시계열 데이터를 2시간 단위로 이동시킨다.
```

```python
crime_dt.shift(-1, freq='D')  # crime_dt의 시계열 데이터를 1일 전씩으로 이동시킨다.
```

```python
crime_dt.shift(1)  # freq 지정 안 할시 데이터 항목의 이동 단위를 의미 (1로 지정하면 한 칸 씩 밀린다.)
```

**5. diff 함수**
- 시계열 데이터의 차분 계산
- 정상성을 띄지 않는 데이터는 차분(differencing)을 통해 시계열 데이터를 변환함으로써 정상성의 특성을 지닌 데이터로 만들어 줘야 AR, MA, ARIMA 등의 전통적인 시계열 분석 방법을 사용할 수 있다.
- 차분을 구한 시계열은 원래 시계열에서 연이은 관측값의 차이이다.

```python
crime['REPORTED_DATE'].diff()  # 시계열 데이터에 해당하는 열을 차분한다
```

**6. rolling 함수**
- 이동 평균이나 이동 합과 같은 이동 통계량을 계산

```python
crime_dt.rolling(window=3).mean()  # crime_dt의 시계열 데이터를 3일 동안의 이동 평균으로 계산
```
