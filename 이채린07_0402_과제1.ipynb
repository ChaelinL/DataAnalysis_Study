**1. append()**
- 변수를 생성해 딕셔너리 타입으로 전체 컬럼에 대응하는 값을 행으로 추가할 수 있다. 

```python
add_data = {'manufacturer':'volkswagen', 'model':'passat', 'displ':'2.8','year':'2009','cyl':'6', 'trans':'auto(s6)', 'drv':'f', 'cty':'21', 'hwy':'30', 'fl':'p','category':'midsize'}
mpg.append(add_data, ignore_index=True)  # 마지막 행에 값이 추가된 것을 확인할 수 있다.
```

- append 함수는 list와 함께 for문에서 유용하게 사용된다!

```python
fruits = ['사과','배','오렌지','레몬']
basket = []
for fruit in fruits:
  basket.append(fruit)
print(basket)
```

**2. describe()**
- Series에 대한 다양한 통계량을 요약해준다.
- DataFrame의 경우 열에 대해 요약이 수행된다.
- 기본적으로 누락데이터(NaN)은 제외되고 데이터 요약이 수행된다.

```python
mpg.describe()
```

**3. drop_duplicates()**
- 중복값이 없는 시리즈를 반환해준다.

```python
mpg.cty.drop_duplicates()
```

**4. get_values()**
- 데이터프레임, 시리즈 등의 객체에서 데이터를 numpy array 형태로 반환한다.

```python
mpg['model'].values
```

**5. isin()**
- 시리즈에 포함된 값이 있는지 확인해준다.
- 파라미터로 list를 받기 때문에 [ ] 안에 찾고자 하는 값을 넣어야 한다.

```python
mpg['manufacturer'].isin(['audi'])
```

**6. mean()**
- 평균값을 알려준다.

```python
mpg.cty.mean()
```

**7. equals()**
- 두 개의 데이터프레임이나 시리즈가 동일한지 비교한다.

```python
df1 = pd.DataFrame({'이름':['홍길동', '김연아', '김사랑'],
                     '나이':[12, 24, 35],
                     '성별':['남','여','여']})

df1
```

```python
df2 = pd.DataFrame({'과일':['수박','사과','레몬'],
                    '색깔':['초록','빨강','노랑']})
df2
```

```python
df1.equals(df2)
```

**8. sort_values()**
- 지정한 순서대로 값을 정렬해준다.
- ascending = True / False를 통해 오름차순과 내림차순을 지정할 수 있다.

```python
mpg.cty.sort_values(ascending=True)  # 값 기반 오름차순
```

**9. to_frame()**
- 시리즈를 데이터 프레임으로 변환시켜준다.

```python
mpg['model'].to_frame()
```
