# 데이터분석 기초반 8주차 수업
- 시계열 데이터 분석
- 시계열이라는 것은 시간의 흐름을 나타낸 데이터라서 시간의 흐름을 바꾸면 문제가 될 수 있다.
- 모델링을 진행할 때도 시계열 데이터는 섞을 수 없다. 시계열 자체의 특성상 섞이게 되면 사라진다.
- 시계열 데이터는 흐름이라는 것이 중요한 특징 중 하나이다.
- 판다스에서 datetime, TimeStamp, Timedelta 등의 함수 제공
- 대부분 datetime을 사용해 데이터 분석을 진행한다.
- 정상성, 차분 등 시계열 분석에 필요한 방법들이 따로 있다.
- 어떠한 경향이나 추세가 보이는 시계열 데이터를 정상시계열로 만든다.

**1. datetime**
- date + time 이 두 가지가 합쳐진 것
- date: 년, 월, 일
- time: 시간, 분, 초, 마이크로초

```python
import pandas as pd
import datetime

date = datetime.date(year=2023, month=5, day=22)
time = datetime.time(hour=22, minute=6, microsecond=1234)

print(date)
print(time)
```

```python
dt = datetime.datetime(year=2023, month=5, day=21, hour=22, minute=6, second=5, microsecond=1234)
print(dt)
```

**2. timedelta**
- 시계열 데이터를 덧셈 또는 뺄셈이 가능하다
- 날짜를 day 계산해서 덧셈과 뺄셈 진행

```python
td = datetime.timedelta(days=20, hours=10)
print(td)
```

```python
print(date+td) # 날짜만 더해서 날짜만 출력
print(dt+td)   # 날짜와 시간 함께 더해서 모두 출력
```

**3. datetime의 장점**
- datetime을 이용해 시계열 데이터로 변경할 수 있다.

```python
pd.to_datetime('2023-5-22')
```

- 날짜 데이터를 표현하는 방식은 다양하다
- 22-5-2023
- 2023/5/22
- 2023.5.22

```python
a = pd.Series(['2023-5-22','22/5/2023','2023-6-21','2023-07-18','2023-05-32'])
pd.to_datetime(a, dayfirst=True, errors='coerce') # 정상적이지 않은 시계열 값은 na 반환
```

- datetime이 반환해주는 형태 중 위 형태는 연도와 월, 일을 스스로 잘 매핑해주며, 잘못된 월 같은 경우 na값을 반환한다.

**4. 실제 데이터를 불러와서 분석해보자!**

```python
# 문자열 데이터를 반드시 시계열 데이터로 꼭 바꿔야 한다.
crime['REPORTED_DATE'] = pd.to_datetime(crime['REPORTED_DATE'])
```

```python
crime.info()  # 시계열 데이터로 바뀌었는지 확인
```

- 해당 시계열 데이터 컬럼을 인덱스로 지정하게 되면 다양하게 사용할 수 있다.
- 시계열 데이터 컬럼을 인덱스로 지정하면 좋은 점: 다양한 메서드를 사용할 수 있다.

```python
crime_dt = crime.set_index('REPORTED_DATE')
```

```python
crime_dt.loc['2014']  # 2014년 데이터 불러오기
```

```python
crime_dt.loc['2017']
```

```python
# 추가로 요일을 더 지정해서 불러올 수도 있다
crime_dt.loc['2017-09']
```

```python
crime_dt.loc['2017-09-13']
```

```python
# 이렇게 불러오는 것도 가능하다
crime_dt.loc['Dec 2014']
```

```python
# sort_index() 사용해 인덱스 순서대로 정렬 가능
crime_dt.loc['2014'].sort_index()
```

**5. between_time**
- 특정 두 시간대 안에 속하는 데이터를 불러온다

```python
crime_dt14 = crime_dt.loc['2014'].sort_index()
```

```python
crime_dt14.between_time('13:00', '17:00')  # 시간을 이용하여 추출할 수도 있다.
```

**6. at_time**
- 해당 시간 정각의 데이터를 불러온다

```python
crime_dt14.at_time('16:49')
```
