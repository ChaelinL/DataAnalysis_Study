```python
import numpy as np
import pandas as pd
```

```python
# 데이터 불러오기
from google.colab import drive
drive.mount('/content/drive')

mpg = pd.read_csv('/content/drive/My Drive/mpg.csv')
mpg
```

Q1. 고속도로연비가 30 이상이면서 도시연비가 20 이상인 자동차들의 제조사는?

```python
mpg.query('hwy>=30 & cty>=20')['manufacturer']
```

Q2. 고속도로연비와 도시연비가 높은 순으로 데이터를 정렬하면?

```python
mpg.sort_values(['hwy','cty'], ascending=False)
```

Q3. 고속도로연비와 도시연비의 평균을 각각 모델을 기준으로 표로 나타내면?

```python
mpg.groupby('model').agg(mean_hwy=('hwy', 'mean'),
                         mean_cty=('cty','mean'))
```

Q4.고속도로연비와 도시연비의 평균을 나타내는 열을 나타내면?

```python
mpg.assign(mean=(mpg['hwy']+mpg['cty'])/2)
```

Q5. 표에 mean이라는 항목을 아예 추가하고, 이를 내림차순으로 정렬한다면?

```python
mpg['mean'] = ((mpg['hwy']+mpg['cty'])/2)
mpg.sort_values('mean', ascending=False)
```

Q6. 고속도로연비와 도시연비의 평균이 30 이상일때 good라고 나타내면?

```python
mpg.assign(mileage=np.where(mpg['mean']>=30, 'good', 'normal'))
```

Q7. 모델별 최고의 연비평균은 무엇인가?

```python
mpg.groupby('model').agg(best_mileage=('mean', 'max'))
```

Q8. 출시년도가 가장 최신인 모델을 기준으로 평균 연비를 나타내면?

```python
# mpg.sort_values('year', ascending=False)
mpg.query('year==2008')[['model','mean']]
```

Q9. 연비평균이 30이상인 모델과 제조사들을 나타내면?

```python
mpg.query('mean>=30')[['manufacturer', 'model']]
```

Q10. 출시년도가 최근인 것을 기준으로 연비평균이 30이상일 때, 모델과 제조사들을 나타내면?

```python
mpg.sort_values(['year'], ascending=False).query('mean>=30')[['manufacturer', 'model']]
```
