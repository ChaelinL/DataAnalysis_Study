- scatter, plot 동일하게 사용되는 것도 있고 다른 것도 있다. -> 맞지 않는 것을 사용하면 에러가 나기 때문에 경우에 따라 인자값을 바꿔줘야 한다.
- scatter: x와 y의 분포를 볼 때 정말 많이 사용한다.
데이터 시각화에서 기본적으로 변수간의 관계를 파악하기 위해 본다.
- 비지도 학습시 군집 간의 관계, 차원 축소시 해당 군집이 어떤 식으로 형성되어 있나 볼 때, 군집화하기 전 기본적으로 분포를 볼 때 사용하는 등 다양하게 사용된다!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**임의의 데이터로 scatter 그리기**

```python
# x값과 y값 랜덤 설정

np.random.seed(0)  # 추출 결과 고정
x = np.random.randint(-10,11, size=10)  # -10~10 사이 5개의 정수 임의 추출
y = np.random.randint(-10,11, size=10)  # 10개의 정수 임의 추출

# 그래프 크기 설정
plt.figure(figsize=(8,8))  # 8*8 크기의 그래프 생성

# 그래프 배경 설정
sns.set(style='whitegrid')

# 그래프 제목과 축 제목 설정
plt.title('test')
plt.xlabel('random x')
plt.ylabel('random y')

# 그래프 축 범위 설정
plt.xlim(-10,10)
plt.ylim(-10,10)

# 눈금값 조절
plt.xticks([-10, -5, 0, 5, 10])
plt.yticks([-10, -5, 0, 5, 10])

# 그래프 그리기 
plt.scatter(x,y, label='example', marker='*', color='red', alpha=0.5)  # marker 설정, color 설정, 투명도 설정
plt.legend()  # 범례 설정
```

**동일한 방법으로 plot도 그릴 수 있다.**

```python
# 그래프 크기 설정
plt.figure(figsize=(8,8))  # 8*8 크기의 그래프 생성

# 그래프 제목과 축 제목 설정
plt.title('test')
plt.xlabel('random x')
plt.ylabel('random y')

# 그래프 축 범위 설정
plt.xlim(-10,10)
plt.ylim(-10,10)

# 눈금값 조절
plt.xticks([-10, -5, 0, 5, 10])
plt.yticks([-10, -5, 0, 5, 10])

# 그래프 그리기
plt.plot(x,y, label='example', marker='o', markersize=5, color='orange', linestyle='dashed', linewidth=5)  # marker 설정, color 설정, line 설정
plt.legend()
```
