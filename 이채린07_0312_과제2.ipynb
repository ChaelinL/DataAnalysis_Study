```python
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기
import seaborn as sns
df = sns.load_dataset('titanic')
df
```

Q1. 생존한 사람들의 객실등급을 나타내면?

```python
df.query('survived==1')['class']
```

Q2. 3등석에 탑승했음에도 생존한 사람들은?

```python
df.query('survived==1 & pclass==3')
```

Q3. 형제부모를 포함한 가족과 함께 탑승한 사람들을 표에 나타내면?

```python
df.assign(family=(df['sibsp']+df['parch']))
```

Q4. 객실등급별 평균 탑승요금은?

```python
df.groupby('pclass').agg(avg_fare=('fare','mean'))
```

Q5. 객실등급별 제일 비싼 탑승요금과 제일 싼 탑승요금을 비교해보면?

```python
df.groupby('pclass').agg(max_fare=('fare', 'max'),
                         min_fare=('fare', 'min'))
```

Q6. 탑승요금이 300이 넘는 사람들을 구분해서 표에 나타내면?

```python
df.assign(money=np.where(df['fare']>300, 'rich', 'normal'))
```

Q7. 생존자 여부와 요금 평균을 비교하면?

```python
df.groupby('survived').agg(avg_fare=('fare','mean'))
```

Q8. 객실등급이 낮고 나이가 어린 순으로 정렬한다면?

```python
df.sort_values(['pclass', 'age'], ascending=[False, True])
```

Q9. 3등칸 탑승 생존자들의 성별과 나이를 나타내면?

```python
df.query('pclass==3 & survived==1')[['sex','age']]
```

Q10. 3등칸 탑승 생존자 중 나이가 12살보다 어리고, 가족이 있는 경우는?

```python
df['family'] = df['sibsp'] + df['parch']
df.query('pclass==3 & survived==1 & age<12')['family']
```
