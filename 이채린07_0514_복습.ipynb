# 데이터분석 기초반 7주차 수업
- 데이터 병합 merge, concat

```python
import pandas as pd
```

**concat 함수**
- df+ df끼리 합치는 경우
- 두 테이블이 행과 열 기준으로 붙여진다.
- ignore_index: 병합 후 인덱스 새롭게 만들 지 선택 (True일시 0부터 다시 시작)
- axis: 축 설정 (axis=0이 디폴트, axis=1는 cbind 형식)
- 컬럼이 동일한 경우 쉽게 붙일 수 있다.
- 단, 시계열 데이터는 다를 수 있으니 유의하기!

```python
df1 = pd.DataFrame({'A':[1,2,3,4], 'B':[1,2,3,4]})
df2 = pd.DataFrame({'A':[3,4], 'B':[4,5]})
df3 = pd.DataFrame({'A':[5,6], 'C':[3,2]})

display(df1)
display(df2)
display(df3)
```

```python
pd.concat([df1, df2, df3], ignore_index=True) # axis=0이 디폴트값
```

```python
pd.concat([df1, df2, df3], axis=1) # cbind 형식으로 병합
```

**merge 함수**
- 두 개의 데이터프레임을 join하는 함수
- 공통된 열을 기준으로 데이터프레임을 병합하는 것이 가능하다.
- 공통된 열(컬럼)이 join key!
- 함수 주요 인자:
- left: 병합할 첫 번째 데이터프레임으로, 왼쪽
- right: 별합할 두 번째 데이터프레임으로, 오른쪽
- on: 조인 키로 사용할 열 목록으로, 두 데이터프레임 모두 포함되어야 한다.
- left_on: 왼쪽 데이터프레임에서 조인키로 사용할 열 목록
- right_on: 오른쪽 데이터프레임에서 조인키로 사용할 열 목록
- left_index: 왼쪽 데이터프레임의 조인키로 인덱스를 사용할 지
- right_index: 오른쪽 데이터프레임의 조인키로 인덱스를 사용할 지
- how: 조인의 방법을 결정하며, 디폴트는 'inner'

```python
df1 = pd.DataFrame({'이름':['이채린','임하나','김민서','윤다은'],
                    '영어':[100,50,60,70]})
df2 = pd.DataFrame({'이름':['이채린','임하나','김민서'],
                    '수학':[50,100,60]})

display(df1)
display(df2)
```

```python
pd.merge(df1, df2) # 디폴트값은 inner 교집합
```

```python
pd.merge(df1, df2, on='이름')  # '이름'열을 기준으로 병합
```

```python
df3 = pd.DataFrame({'이름':['이채린','임하나','김민서','윤다은'],
                    '수학':[50,100,60,70],
                    '생년월일':['00','01','02','03']})
df4 = pd.DataFrame({'이름':['이채린','임하나','김민서'],
                    '수학':[50,100,60],
                    '지각횟수':[0,1,2]})

display(df3)
display(df4)
```

```python
pd.merge(df3, df4, on=['이름','수학'])  # on을 사용할 때 2개 이상 컬럼도 사용 가능
```

```python
pd.merge(df3, df4, on='이름')  # on='수학'하지 않으면 두 개로 나뉘어 나온다.
```

```python
# left_on, right_on

df5 = pd.DataFrame({'이름':['이채린','임하나','김민서','윤다은'],
                    '수학':[50,100,60,70],
                    '생년월일':['00','01','02','03']})
df6 = pd.DataFrame({'name':['이채린','임하나','김민서'],
                    '수학':[50,100,60],
                    '지각횟수':[0,1,2]})

display(df5)
display(df6)
```

```python
pd.merge(df5, df6, left_on='이름', right_on='name')  # left_on과 right_on은 같이 사용되어야 함! (or left_index나 right_index도 가능)
```

```python
pd.merge(df5, df6, left_on=['이름','수학'], right_on=['name','수학'])  # left_on과 right_on의 길이도 같아야 함!
```

```python
# left_index, right_index

df5_sp = df5.set_index('이름')
display(df5_sp)
display(df6)
```

```python
pd.merge(df5_sp, df6, left_index=True, right_on='name')
```

**how**
- 조인키에 대한 기본값 = inner
- left: 왼쪽 데이터프레임 조인키를 기준으로 병합
- right: 오른쪽 데이터프레임 조인키를 기준으로 병합
- outer: 왼쪽과 오른쪽 데이터프레임 조인키에 대한 합집한 기준으로 병합

```python
df1 = pd.DataFrame({'A':[1,2,3], 'B':[1,2,3]})
df2 = pd.DataFrame({'A':[2,3,4], 'C':[7,8,9]})

display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on='A', how='inner')
```

```python
pd.merge(df1, df2, on='A', how='right')
```

```python
pd.merge(df1, df2, on='A', how='left')
```

```python
pd.merge(df1, df2, on='A', how='outer')
```
